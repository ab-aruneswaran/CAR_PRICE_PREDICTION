```python
make_train = X_train['CarName'].str.split(" ",expand =  True)
make_test = X_test['CarName'].str.split(" ",expand = True)
X_train['CarName'] = make_train[0]
X_test['CarName'] = make_test[0]
```

```python
plt.figure(figsize = (12, 8))
plot = sns.countplot(x = 'CarName', data = X_train)
plt.xticks(rotation = 90)
for p in plot.patches:
    plot.annotate(p.get_height(),(p.get_x() + p.get_width()/2.0,p.get_height()),
    ha = 'center',
    va = 'center',
    xytext = (0,5),
    textcoords = "offset points")
plt.title("Counts of cars based on Manufacturers")
plt.xlabel("Manufacturers")
plt.ylabel("Count of Cars")
```

```python
predict = "price"
df = df[["symboling", "wheelbase", "carlength", 
             "carwidth", "carheight", "curbweight", 
             "enginesize", "boreratio", "stroke", 
             "compressionratio", "horsepower", "peakrpm", 
             "citympg", "highwaympg", "price"]]
x = np.array(df.drop([predict], 1))
y = np.array(df[predict])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

Model Selection

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(xtrain, ytrain)
```

```python
y_pred = model.predict(xtest)
```

Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
```

```python
model.score(xtest, y_pred)
```
